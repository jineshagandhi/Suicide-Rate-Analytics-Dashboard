```python
import pandas as pd

def preprocess_merged_data(input_csv, output_csv):
    """
    Preprocesses the merged final dataset by:
    - Removing non-state summary rows (like 'India')
    - Stripping whitespace from state names
    - Ensuring correct numeric types
    - Filling missing values with sensible defaults
    - Recalculating suicide rate per lakh
    
    Args:
        input_csv (str): Path to the merged final dataset.
        output_csv (str): Path to save the cleaned dataset.
    """
    # 1. Load the dataset
    df = pd.read_csv(input_csv)

    # 2. Remove rows with 'India' or null in State
    df = df[df['State'].notna()]
    df = df[df['State'].str.lower() != 'india']

    # 3. Strip whitespace from State names
    df['State'] = df['State'].str.strip()

    # 4. Convert numeric columns
    numeric_cols = ['Literacy_Total', 'Literacy_Male', 'Literacy_Female',
                    'Unemployment_Rate', 'Total_Suicides', 'Population']
    for col in numeric_cols:
        df[col] = pd.to_numeric(df[col], errors='coerce')

    # 5. Fill NaN values
    df['Unemployment_Rate'] = df['Unemployment_Rate'].fillna(0)
    df['Total_Suicides'] = df['Total_Suicides'].fillna(0)
    df['Population'] = df['Population'].fillna(df['Population'].median())

    # 6. Recalculate suicide rate per lakh
    df['Suicide_Rate_Per_Lakh'] = (df['Total_Suicides'] / df['Population']) * 100000

    # 7. Save cleaned file
    df.to_csv(output_csv, index=False)
    print(f"✅ Preprocessed dataset saved to '{output_csv}'")

# Example usage
if __name__ == "__main__":
    preprocess_merged_data(
        input_csv='merged_final_data.csv',
        output_csv='merged_final_data_cleaned.csv'
    )
```
