```python
import pandas as pd
import numpy as np
```

```python
literacy_df = pd.read_csv("../data/raw/literacy.csv")
unemployment_df = pd.read_csv("../data/raw/unemployment.csv")
suicide_df = pd.read_csv("../data/raw/suicides.csv")
```

```python
literacy_df = literacy_df[['State/UT', 'Total', 'Male', 'Female']]
literacy_df.columns = ['State', 'Literacy_Total', 'Literacy_Male', 'Literacy_Female']
literacy_df['State'] = literacy_df['State'].str.strip()
```

```python
unemployment_df = unemployment_df[['State/Union Territory', '2011-12']]
unemployment_df.columns = ['State', 'Unemployment_Rate']
unemployment_df['State'] = unemployment_df['State'].str.strip()
```

```python
suicide_df = suicide_df[suicide_df['Year'] == 2011]
suicide_statewise = suicide_df.groupby('State', as_index=False)['Total'].sum()
suicide_statewise.columns = ['State', 'Total_Suicides']
suicide_statewise['State'] = suicide_statewise['State'].str.strip()
```

```python
state_mapping = {
    "A & N Islands": "Andaman and Nicobar Islands",
    "Andaman and Nicobar Islands[UT]": "Andaman and Nicobar Islands",
    "NCT of Delhi": "Delhi",
    "Delhi[UT]": "Delhi",
    "Jammu & Kashmir": "Jammu and Kashmir",
    "Jammu & Kashmir[UT]": "Jammu and Kashmir",
    "Telangana": "Telangana",
    "Puducherry[UT]": "Puducherry",
    "Dadra & Nagar Haveli": "Dadra and Nagar Haveli and Daman and Diu",
    "Daman & Diu": "Dadra and Nagar Haveli and Daman and Diu",
    "Ladakh[UT]": "Ladakh",
    "Arunachal Pradesh": "Arunachal Pradesh",
    "Uttaranchal": "Uttarakhand",
    "Orissa": "Odisha"
}

def clean_state_column(df, col):
    return df[col].str.strip().replace(state_mapping)

literacy_df["State"] = clean_state_column(literacy_df, "State")
unemployment_df["State"] = clean_state_column(unemployment_df, "State")
suicide_statewise["State"] = clean_state_column(suicide_statewise, "State")
```

```python
merged_df = pd.merge(literacy_df, unemployment_df, on='State', how='inner')
merged_df = pd.merge(merged_df, suicide_statewise, on='State', how='inner')
```

```python
merged_df.to_csv("final_dataset_for_tableau.csv", index=False)

```
